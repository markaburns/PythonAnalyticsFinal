```python
import pandas
from matplotlib import pyplot
spotifydf=pandas.read_csv('https://raw.githubusercontent.com/markaburns/PythonAnalyticsFinal/main/spotifydataset_truncated.csv')

"""
Examine the degree of correlation between popularity and duration ... does popularity roll off with duration?
"""

y=spotifydf["popularity"]
x=spotifydf["duration_ms"]
fig, ax=pyplot.subplots()
ax.set_ylabel("Popularity per Spotify Algorithm (0-100)")
ax.set_xlabel("Song Duration (in mSec)")
ax.set_title("Correlation Check of Popularity vs. Duration")
ax.scatter(x,y)
pyplot.show()

# Correlation b/w Cases vs Deaths (cumulative)

correlation = spotifydf['duration_ms'].corr(spotifydf['popularity'])
print('Correlation between song duration and popularity:',{correlation})

# Is the correlation significant? (Correlationis significant if its absolute value is greater than 0.75)

if abs(correlation) > 0.75:
    print("Correlation is significant.")
else:
    print("Correlation is not significant.")
```

```python
import pandas
from matplotlib import pyplot
spotifydf=pandas.read_csv('https://raw.githubusercontent.com/markaburns/PythonAnalyticsFinal/main/spotifydataset_truncated.csv')

"""
Examine the degree of correlation between popularity and danceability
"""

y=spotifydf["popularity"]
x=spotifydf["danceability"]
fig, ax=pyplot.subplots()
ax.set_ylabel("Popularity per Spotify Algorithm (0-100)")
ax.set_xlabel("Danceability per Spotify Algorithm (0-1.0)")
ax.set_title("Correlation Check of Popularity vs. Danceability")
ax.scatter(x,y)
pyplot.show()


correlation = spotifydf['danceability'].corr(spotifydf['popularity'])
print('Correlation between danceability and popularity:',{correlation})

# Is the correlation significant? (Correlationis significant if its absolute value is greater than 0.75)

if abs(correlation) > 0.75:
    print("Correlation is significant.")
else:
    print("Correlation is not significant.")
```
